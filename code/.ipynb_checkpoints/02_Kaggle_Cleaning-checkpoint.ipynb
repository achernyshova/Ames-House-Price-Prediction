# Kaggle Dataset Cleaning

This notebook contains data cleaning for test dataset <br>
I did the exact same cleaning manipulations as on the training data. 

## Import Libriaries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

pd.set_option('display.max_rows', 1000) 
pd.set_option('display.max_columns', 500) 
pd.set_option('display.width', 1000)
```

## Load the Data

```python
df = pd.read_csv('../datasets/test.csv')
```

## Data Cleaning

```python
df.head()
```

```python
df.shape
```

**Missing Values:**

```python
df.isnull().sum().sort_values(ascending=False)[:30]
```

```python
categotical_columns = ['Pool QC', 'Misc Feature', 'Alley', 'Fence', 'Fireplace Qu', 'Garage Finish', 'Garage Cond', 
              'Garage Qual', 'Garage Type', 'Bsmt Exposure', 'BsmtFin Type 2', 'BsmtFin Type 1', 'Bsmt Cond', 
              'Bsmt Qual']
```

```python
[df[x].replace(np.nan, 'NA', inplace=True) for x in categotical_columns]
df.isnull().sum().sort_values(ascending=False)[:15]
```

```python
df['Mas Vnr Type'].replace(np.nan, 'None',inplace=True )
```

```python
numeric_columns = ['Garage Yr Blt', 'Mas Vnr Area','Bsmt Half Bath', 'Bsmt Full Bath', 'Garage Cars', 'Garage Area',
                   'Bsmt Unf SF', 'BsmtFin SF 2', 'Total Bsmt SF', 'BsmtFin SF 1']
```

```python
[df[x].replace(np.nan, 0 , inplace=True) for x in numeric_columns]
df.isnull().sum().sort_values(ascending=False)[:10]
```

```python
df['Lot Frontage'] = df.groupby('Neighborhood')['Lot Frontage'].transform(lambda x: x.fillna(x.median()))
df['Lot Frontage'].isnull().sum()
```

```python
df['Electrical'].replace(np.nan, df['Electrical'].mode() ,inplace=True )
```

**Data types**

```python
df.dtypes
```

**Rename columns**

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.columns
```

**Export clean data to csv**

```python
df.to_csv('../datasets/kaggle_clean.csv')
```
